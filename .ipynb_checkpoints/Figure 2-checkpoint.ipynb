<h1>Figure 2</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

<h2>Data sources</h2>
<p>Data for SSU footprints on yeast mRNAs were originally described in <a href="https://www.ncbi.nlm.nih.gov/pubmed/27437580" target=_blank>Archer <i>et al.</i> 2016</a>. The SSU track for this publication was downloaded via the table browser function in <a href="https://gwips.ucc.ie/index.html" target=_blank>GWIPS-viz</a>. Matching gene coordinates were downloaded from <a href="https://www.yeastgenome.org/download-data" target=_blank>SGD</a> (sacCer3).</p>

<h2>Data processing</h2>
<p>Coverage data corresponding to the region -200/+50 nucleotides around the start codon of each gene were extracted and saved to python dictionary format using script 'track_to_dict.py'. Peaks within this region were called using script 'dict_to_peaks.py', which is based on the 'find_peaks_cwt' function from scipy.signal.</p>

<h2>SSU footprint file load</h2>

<p>The cell below opens the pickled dictionary containing the -200/+50 traces of SSU footprints, and the pickled dictionary of peak locations within these footprints.</p>

```python
with open(r"_Data/An_UTR_SSU_prints.p", "rb") as input_file:
    footprint_dict = pickle.load(input_file)
    
with open(r"_Data/An_UTR_SSU_peaks.p", "rb") as input_file:
    peaks_dict = pickle.load(input_file)
```

<h2>Plot SSU peaks for individual genes</h2>
<p>The cell below can be used to print plots of the footprint coverage for individual genes, with identified peak locations indicated. To display different genes change the value of query_term in the first row (accepts standard or systematic gene names).</p>

```python
query_term = 'HIS3'

coordinates = pd.read_csv('_Data/genome_sacCer3_genes.tsv', sep='\t')

#determine which gene the query term corresponds to

if query_term.upper() in coordinates['name'].values:
    query_gene=query_term.upper()
    show=True
elif query_term.upper() in coordinates['proteinID'].values:
    query_gene=coordinates.loc[coordinates['proteinID']==query_term.upper()]['name']
    query_gene = query_gene.reset_index(drop=True)[0]
    show=True
else: 
    print('no data')
    show=False

if show:
#determine the peaks for this gene
    peak_x = peaks_dict[query_gene]
    peak_y = footprint_dict[query_gene].flatten()[peak_x+200]

#produce the plot
    data = footprint_dict[query_gene].flatten()
    plt.plot(np.linspace(-199,50,250), data)
    plt.scatter(peak_x,peak_y, marker="+",color="red",s=200)
    plt.title(query_term.upper())
    #plt.savefig(query_term + '_SSU.svg')
    plt.show()
else:
    plt.close()
    
    
```

<h2>5'-UTR lengths and second peaks</h2>
<p>The following cell reads in the speed data table generated in previous anlayses and adds information on the 5'-UTR lengths based on data from the publications by <a href="https://www.ncbi.nlm.nih.gov/pubmed/17101987" target=_blank> Miura <i>et al.</i> 2006</a>, <a href="https://www.ncbi.nlm.nih.gov/pubmed/18451266" tagret=_blank>Nagalakshmi <i>et al.</i> 2008</a>, and <a href="https://www.ncbi.nlm.nih.gov/pubmed/23390610" target=_blank>Waern and Snyder 2013</a> (using the longest reported 5'-UTR in any of these publications). It also adds a column to the speed data indicating whether a second SSU peak is present in the region between 15 and 60 nt upstream of the start codon.</p>

```python
#import the data tables containg 5'-UTR information
UTR_table = pd.read_csv('_Data/An_firstTenNew.csv')
Miura = pd.read_csv('_Data/Lit_Miura_2006_5_UTRs.csv')
Naga = pd.read_csv('_Data/Lit_Nagalakshmi_2008_5_UTRs.csv')
Waern = pd.read_csv('_Data/Lit_Waern_Snyder_2013_5_UTR.csv')

#look up the 5'-UTR annotations for each gene and record the highest value
UTRs = []

for Gene in UTR_table['ORF']:
    M5,N5,W5=0,0,0
    if Gene in Miura['ORF'].values:
        M5 = int(max(Miura['5_UTR'].loc[Miura['ORF']==Gene].values))
    if Gene in Naga['ORF'].values:
        N5 = int(max(Naga['5_UTR'].loc[Naga['ORF']==Gene].values))
    if Gene in Waern['ORF'].values:
        W5 = int(max(Waern['5_UTR'].loc[Waern['ORF']==Gene].values))
    UTRs.append(max([M5,N5,W5]))


UTR_table['5_UTR'] = UTRs
```

```python
#for each gene, look up whether a second peak is present in the area between -60 and -15 nt
second_SSU_peak = []
for Gene in UTR_table['ORF']:
    append_value = False
    if Gene in peaks_dict.keys():
        if len(list(peaks_dict[Gene])) >= 2:
            if list(peaks_dict[Gene])[-2] >= -60:
                if list(peaks_dict[Gene])[-2] <= -15:
                    append_value = True
    second_SSU_peak.append(append_value)
                    
UTR_table['second_SSU_peak'] = second_SSU_peak
```

```python
UTR_table.to_csv('_Data/An_firstTenNew_SSU.csv')
```

<h2>Meta gene plots of SSU footprints</h2>
<p>This cell reads in the speed data table generated in the previous anlayses, and divides all genes where the 5'-UTR length is >45 nt (ie where a waiting SSU can be observed in principle) into two classes depending on whether or not they contain a waiting SSU peak. It then prepares and plots meta gene plots for the SSU coverage of each class. This cell requires running the "5'-UTRs" cell first.</p>

```python
#discard all genes where the 5'-UTR is too short to accomodate a second SSU (<45 nt)
master_table = UTR_table.loc[UTR_table['5_UTR'] >= 45]
#make a list of those genes which are positive for a second SSU peak
second_peak_list = list(master_table['ORF'].loc[master_table['second_SSU_peak']])
meta_gene_array=np.zeros(250)
#append all arrays for second peak genes into one two-dimensional array
counter = 0
for gene in second_peak_list:
    meta_gene_array = np.concatenate((meta_gene_array,footprint_dict[gene].flatten()), axis=0) 
    counter = counter + 1

meta_gene_array = meta_gene_array.reshape((counter + 1,250))
#remove the first array of zeros
meta_gene_array = meta_gene_array[1:,:]
#calculate the metagene footprints as the mean of the individual footprints
meta_gene_twoprints = np.mean(meta_gene_array, axis = 0)

#now process all genes with only a single peak
no_second_SSU_peak = [not i for i in master_table['second_SSU_peak']]
no_second_peak_list = list(master_table['ORF'].loc[no_second_SSU_peak])
meta_gene_array=np.zeros(250)
#append all arrays for non-second peak genes into one two-dimensional array
counter = 0
for gene in no_second_peak_list:
    #check if the gene exists in the Archer data set
    if gene in footprint_dict.keys():
        meta_gene_array = np.concatenate((meta_gene_array,footprint_dict[gene].flatten()), axis=0) 
        counter = counter + 1

meta_gene_array = meta_gene_array.reshape((counter + 1,250))
#remove the first array of zeros
meta_gene_array = meta_gene_array[1:,:]
#calculate the metagene footprints as the mean of the individual footprints
meta_gene_oneprint = np.mean(meta_gene_array, axis = 0)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(np.linspace(-199,50,250),meta_gene_oneprint, color='orange')
Class_I_title = "Class I (n=" + str(len(no_second_peak_list)) + ")"
ax1.set_title(Class_I_title)
ax1.set_xlim(-120,50)
ax2.plot(np.linspace(-199,50,250),meta_gene_twoprints, color='blue')
Class_II_title = "Class II (n=" + str(len(second_peak_list)) + ")"
ax2.set_title(Class_II_title)
ax2.set_xlim(-120,50)
#plt.savefig('SSU_metagenes.svg')
plt.show()
```

##### <h2>Mapping of "second peak" genes onto the volcano plot</h2>
<p>This cell reproduces the volcano plot for the speed data for those genes which have an annotated 5'-UTR length > 45 nt, colour coded by whether or not they contain a waiting SSU peak.</p>

```python
#discard all genes where the 5'-UTR is too short to accomodate a second SSU (<45 nt)
dataset = UTR_table.loc[UTR_table['5_UTR'] > 45]
#split the dataset into two depending on whether or not a waiting SSU is detected
second_SSU_peak = dataset['second_SSU_peak']
no_second_SSU_peak = [not i for i in dataset['second_SSU_peak']]
orange_set = dataset.loc[no_second_SSU_peak]
blue_set = dataset.loc[second_SSU_peak]

#generate plot
plt.scatter(orange_set['DecodingTime'], orange_set['P'],color='orange',alpha=0.4,s=3)
plt.scatter(blue_set['DecodingTime'], blue_set['P'],color='blue',alpha=0.6,s=3)
plt.xlabel('$Decoding\ Time\ (first\ ten\ codons)\ [sec]$')
plt.ylabel(r'$Probability\ [log_{10}]$')
#plt.savefig('Volcano_SSU.svg')
plt.show()
```

```python
proportion = blue_set.shape[0]/dataset.shape[0]*100
print("Out of " + str(dataset.shape[0]) + " genes with a 5'-UTR length > 45 nt, " + str(dataset.loc[dataset['second_SSU_peak']].shape[0]) + " genes contain a second peak indicating the presence of waiting small ribosomal subunits.")
print("%.1f" % proportion + "% of yeast genes with a 5'-UTR length > 45 nt contain a second SSU peak")
```
